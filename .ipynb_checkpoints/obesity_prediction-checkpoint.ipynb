## Data Preprocessing

### Encoding Categorical features

```python
# Encoding categorical features 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0,4,5,8,9,11,14,15])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

### Encoding Categorical Label

```python
# Encoding categorical label
def encode_label(category):
    if category == "Insufficient_Weight":
        return 1
    elif category == "Normal_Weight":
        return 2
    elif category == "Overweight_Level_I":
        return 3
    elif category == "Overweight_Level_II":
        return 4
    elif category == "Obesity_Type_I":
        return 5
    elif category == "Obesity_Type_II":
        return 6
    else: return 7

vectorized_encode_label = np.vectorize(encode_label) ## I vectorized the function so that it can work with numpy arrays
```

```python
y = vectorized_encode_label(y)
```

## Splitting Data into training and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

## Feature Scaling

scaling the age, height, and weight to prevent outliers

```python
print(X_test[:, 23:26])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 23:26] = sc.fit_transform(X_train[:, 23:26])
X_test[:, 23:26] = sc.transform(X_test[:, 23:26])
```

```python
print(X_test[:, 23:26])
```

## Training the Machine Learning Model (Regression)

### General function for calculating the metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def compute_metrics(y_pred):
    # Display the results
    print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))
    print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
    print("R² Score:", r2_score(y_test, y_pred))
```

### Training using multiple linear regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
compute_metrics(lin_reg.predict(X_test))
```

### Training using support vector regression

```python
from sklearn.svm import SVR 

sv_reg = SVR(kernel='rbf')
sv_reg.fit(X_train, y_train)
```

```python
compute_metrics(sv_reg.predict(X_test))
```

### Training using Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
```

```python
compute_metrics(dt_reg.predict(X_test))
```

### Training using Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()

rf_reg.fit(X_train, y_train)
```

```python
compute_metrics(rf_reg.predict(X_test))
```

## Conclusion

From our algorithms, the model with the best accuracy is the random forest regression model having an r2 score of 0.99
